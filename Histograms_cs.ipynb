```python
import pandas as pd
# h1b_raw_2017 = pd.read_csv("data/H-1B_Disclosure_Data_FY17.csv", low_memory=False)
h1b_raw_2018 = pd.read_csv("data/H-1B_Disclosure_Data_FY2018_EOY.csv", low_memory=False)
h1b_raw_2019 = pd.read_csv("data/H-1B_Disclosure_Data_FY2019.csv", low_memory=False)
```

```python
all_raw_data_ltd = h1b_raw_2019[['CASE_STATUS', 'SOC_CODE', 'FULL_TIME_POSITION']].\
                    append(h1b_raw_2018[['CASE_STATUS', 'SOC_CODE', 'FULL_TIME_POSITION']])
all_raw_data_ltd.dropna(subset=['CASE_STATUS'], inplace=True)
```

```python
all_raw_data_ltd.head(5)
```

```python
all_raw_data_ltd['CASE_STATUS'].value_counts(dropna=False)
# h1b_raw_2019['CASE_STATUS'].value_counts(dropna=False)
# h1b_raw_2019[h1b_raw_2019['CASE_STATUS'].isna()].head(10)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,7))
plt2 = pd.DataFrame(all_raw_data_ltd['CASE_STATUS'].value_counts())
ax.bar(x=plt2.index, height=plt2['CASE_STATUS'])
ax.tick_params(axis='both', which='major', labelsize=10)
plt.suptitle('Histogram of Case Status (Outcome)', fontsize=20)
plt.ylabel('# of applications', fontsize=12)
plt.show()
```

```python
all_raw_data_ltd2 = all_raw_data_ltd[~all_raw_data_ltd['CASE_STATUS'].str.contains('WITHDRAWN')]
```

```python
fig, ax = plt.subplots(figsize=(10,7))
plt3 = pd.DataFrame(all_raw_data_ltd2['FULL_TIME_POSITION'].value_counts())
ax.bar(x=plt3.index, height=plt3['FULL_TIME_POSITION'])
ax.tick_params(axis='both', which='major', labelsize=15)
plt.suptitle('Histogram of Full-Time Position', fontsize=20)
plt.ylabel('# of applications', fontsize=15)
plt.show()
```
